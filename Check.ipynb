### dfUNet

```python
path1 = "/home/rajesh/Desktop/Masking/dfUNet/Outputs/99/"

dbvocal, fs = lb.load(path1+'bvocal.wav')
dbbass, fs = lb.load(path1+'bbass.wav')
dbdrums, fs = lb.load(path1+'bdrums.wav')
dbother, fs = lb.load(path1+'bother.wav')

dpvocal, fs = lb.load(path1+'pvocal.wav')
dpbass, fs = lb.load(path1+'pbass.wav')
dpdrums, fs = lb.load(path1+'pdrums.wav')
dpother, fs = lb.load(path1+'pother.wav')

dtvocal, fs = lb.load(path1+'tvocal.wav')
dtbass, fs = lb.load(path1+'tbass.wav')
dtdrums, fs = lb.load(path1+'tdrums.wav')
dtother, fs = lb.load(path1+'tother.wav')
```

```python
dbv = lb.stft(dbvocal)
dbb = lb.stft(dbbass)
dbd = lb.stft(dbdrums)
dbo = lb.stft(dbother)

dtv = lb.stft(dtvocal)
dtb = lb.stft(dtbass)
dtd = lb.stft(dtdrums)
dto = lb.stft(dtother)

dpv = lb.stft(dpvocal)
dpb = lb.stft(dpbass)
dpd = lb.stft(dpdrums)
dpo = lb.stft(dpother)
```

```python
dbv_db = lb.amplitude_to_db(np.abs(dbv))
dbb_db = lb.amplitude_to_db(np.abs(dbb))
dbd_db = lb.amplitude_to_db(np.abs(dbd))
dbo_db = lb.amplitude_to_db(np.abs(dbo))

dtv_db = lb.amplitude_to_db(np.abs(dtv))
dtb_db = lb.amplitude_to_db(np.abs(dtb))
dtd_db = lb.amplitude_to_db(np.abs(dtd))
dto_db = lb.amplitude_to_db(np.abs(dto))

dpv_db = lb.amplitude_to_db(np.abs(dpv))
dpb_db = lb.amplitude_to_db(np.abs(dpb))
dpd_db = lb.amplitude_to_db(np.abs(dpd))
dpo_db = lb.amplitude_to_db(np.abs(dpo))
```
