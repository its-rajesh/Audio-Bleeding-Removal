```python
path1 = "/home/rajesh/Desktop/Masking/dfUNet/Outputs/77/"

dbvocal, fs = lb.load(path1+'bvocal.wav')
dbbass, fs = lb.load(path1+'bbass.wav')
dbdrums, fs = lb.load(path1+'bdrums.wav')
dbother, fs = lb.load(path1+'bother.wav')

dpvocal, fs = lb.load(path1+'pvocal.wav')
dpbass, fs = lb.load(path1+'pbass.wav')
dpdrums, fs = lb.load(path1+'pdrums.wav')
dpother, fs = lb.load(path1+'pother.wav')

dtvocal, fs = lb.load(path1+'tvocal.wav')
dtbass, fs = lb.load(path1+'tbass.wav')
dtdrums, fs = lb.load(path1+'tdrums.wav')
dtother, fs = lb.load(path1+'tother.wav')
```
